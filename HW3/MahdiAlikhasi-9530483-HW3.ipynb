```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#import data and preprocess it
dataset = pd.read_csv("Data/TrainData.txt", header=None)
# y = dataset.iloc[:,0].values
X = dataset.iloc[:,:].values

size = (40, 30)
```

```python
# show a sample by matplotlib
def show(z):
    fig = plt.figure(figsize=(12, 12))
    columns = len(z)
    rows = 1

    for i in range(len(z)):
        fig.add_subplot(rows, columns, i + 1)
        img = z[i].reshape(size)
        plt.imshow(img)
    plt.show()

for i in range(7):
    show(X[i * 10:(i+1) *10])
```

```python
def cost(X, z, c):
    sum = 0.0
    for i in range(len(z)):
        d = X[c == i] - z[i]
        d = np.multiply(d, d)
        d = np.sum(d)
        sum += d
    return sum
```

```python
# do K means
def kmean(k, z):
    c = np.zeros(len(X))
    J = cost(X, z , c)
    while True:
        # clustering
        for i in range(len(X)):
            dis = np.zeros(k)
            for j in range(len(z)):
                diff = X[i] - z[j]
                diff = np.dot(diff, diff.transpose())
                dis[j] = diff
            m = dis[0]
            for j in range(len(dis)):
                if(dis[j] < m):
                    m = dis[j]
                    c[i] = j
        
        # find mass center
        for i in range(len(z)):
            z[i] = np.mean(X[c == i], axis=0)
        l = cost(X, z, c)
        if(J - l < 0.1):
            break
        J = l
        #print(J)
    return z, c, J / len(X)
```

```python
# Question 1
k = 10

# part a
z = np.random.rand(k, size[0], size[1])
z = np.random.binomial(1, z)
z = z.reshape(k, 1, -1)
z = z.astype("float64")
z.shape

print("Bordar haye Avalie")
show(z)

fz1a, c, J = kmean(k, z)

print("Bordar haye nahayee")
show(fz1a)
print("Cost : ", J)
print("P.S :: COST == || X - Z || ^ 2")
```
