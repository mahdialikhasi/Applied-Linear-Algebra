```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = [[0, 1.6],[0.3, 0.8]]
a = np.array(a).reshape(2,2)
```

```python
n = 20
solution = []
x0 = np.array([15, 10]).reshape(2,1)
solution.append(x0)
for i in range(0, n):
    t = solution[-1]
    solution.append(np.dot(a, t))
solution = np.array(solution)
```

```python
plt.plot(solution[:, 0])
plt.show()
```

```python
plt.plot(solution[:, 1])
plt.plot()
```

```python
plt.plot(solution[:,0] + solution[:,1])
```

```python
plt.plot(solution[:,0] / solution[:,1])
```
